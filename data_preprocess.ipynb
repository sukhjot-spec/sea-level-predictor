Handling missing values in epa-sea-level.csv

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data/epa-sea-level.csv")
```

```python
df.info()
```

```python
#checking shape of data before deleting 
df.shape
```

```python
#getting total number of null values in the data 
df.isnull().sum()
```

```python
df_cleaned = df.dropna(subset=['CSIRO Adjusted Sea Level'])
```

```python
df_cleaned.shape
```

```python
df_cleaned.to_csv('epa-sea-level-CLEANED.csv', index=False)
```
